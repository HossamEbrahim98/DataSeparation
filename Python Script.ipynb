## Reading Data and Converting it into Data Frame

```python
import pandas as pd
df = pd.read_excel(r'E:\DA Test\Python - Write a script to merge 100 report templates in single accumulated report\Final Report.xlsx')
df.head()
```

## Extract Data for each client in a new data frame 

```python
df1=df.iloc[:,:6]
df1.drop_duplicates(inplace=True)
df1.reset_index(drop=True,inplace=True)
df1.head()
```

### Making two lists :
#### lst for each branch & lst1 for each client name

```python
lst = list(df1['Branch'])
lst1 = list(df1['Merchandiser Name'])
```

# Finally, I splitted the data into two tables and extracted a table with each client data, 
### including the day, date, and branch to which it belongs. 
# In the other table, I put information for each product of the client , 
### including information about the product itself, as well as the availability of this product.
# And repeated this operation for each client

```python
for i in range(len(lst)):
    grouped1 = df.groupby('Branch')
    df_grouped1 = grouped1.get_group(lst[i]).iloc[0,:6]
    df_grouped1['Date']=df_grouped1['Date'].date()
    grouped = df.groupby('Branch')
    df_grouped = grouped.get_group(lst[i]).iloc[:,6:]
    df_grouped.drop_duplicates(inplace=True)
    df_grouped.reset_index(drop=True,inplace=True)
    #print(df_grouped1)
    #print(df_grouped)
    with open('Report.csv','a') as f:
        df_grouped1.to_csv(f,header=False)
        df_grouped.to_csv(f,index=False)
        f.write("\n")
```
